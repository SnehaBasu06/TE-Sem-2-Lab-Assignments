```python
#dropping extra unwanted columns
df2 = df2.drop(['fnlwgt'], axis=1)
df2 = df2.drop(['education.num'], axis=1)
df2 = df2.drop(['capital.gain'], axis=1)
df2 = df2.drop(['capital.loss'], axis=1)
df2.head()
```

```python
#initial statistics
df2.describe()
```

```python
#datatypes of adult.csv
df2.dtypes
```

```python
#changing 'income' datatype to categorical
df2['income'] = df2['income'].astype('category')
df2.dtypes
```

```python
#group analysis of age by income
analysis = df2.groupby('income')['age'].describe()
analysis
```

```python
#group analysis of age by working hours per week
analysis_ = df2.groupby('income')['hours.per.week'].describe()
analysis_
```

```python
#plotting average working hours per week in different age groups
bins =[0,20,30,40,50,60,70,80,90,100]
labels = ['0-20','21-30','31-40','41-50','51-60','61-70','71-80','81-90','91-100']
df2['age_range']= pd.cut(df2['age'], bins=bins,labels=labels,right = False)
grouped_age = df2.groupby('age_range')['hours.per.week'].mean().reset_index()
plt.bar(grouped_age['age_range'],grouped_age['hours.per.week'],color='teal')
plt.xlabel('Age Range')
plt.ylabel('Avg Working hours per week')
plt.title('Average Working Hours per week across age Ranges')
plt.show()
```

```python
#converting workclass to ctaegoricadf2['workclass']=df2['workclass'].astype('category')
df2.dtypes
```

```python
#plotting avg working hours in different workclass
fig = plt.figure(figsize=(15,3))
grouped_workclass = df2.groupby('workclass')['hours.per.week'].mean().reset_index()
plt.bar(grouped_workclass['workclass'],grouped_workclass['hours.per.week'],color='pink')
plt.xlabel('Working Class')
plt.ylabel('Avg Working hours per week')
plt.title('Average Working Hours per week across Countries')
plt.show()
```
